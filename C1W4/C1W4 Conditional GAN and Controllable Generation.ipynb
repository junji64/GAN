## Conditional GAN & Controllable Generation
지난 몇 주 동안 GAN이 작동하는 방식과 훈련 데이터 세트를 모방한 예제를 생성하기 위해 GAN을 구축하는 방법을 살펴보았습니다. 이번 주에는 출력을 제어하고, 특정 클래스의 예제를 만들거나, 해당 예제에서 특정한 특징을 가지도록 하는 방법을 보여 드리겠습니다. 이것은 꽤 재미있습니다. 이 비디오에서 우선 무조건 생성이 무엇인지 검토할 것입니다. 사실 이것은 여러분이 지금까지 내내 사용하고 있는 것이고, 조건적 생성에 대해서 소개하며, 이 둘을 비교해 보겠습니다.

* Unconditional generation
* 조건적(conditional) vs. 무조건(unconditional) generation

### Conditional generation: Intuition
**무조건 생성**에 대한 간략한 요약으로 임의의 클래스에서 출력을 얻습니다. **동전을 입력하면 임의의 색상의 검볼을 얻는 검볼 기계**로 생각할 수 있습니다. 빨간색과 같이 특정 색상의 검볼을 얻고 싶다면 얻을 때까지 동전을 계속 넣어 돌려야 합니다. 이 예에서 **코인은 GAN이 생성에 사용하는 랜덤 노이즈 벡터**와 같고 **검볼 머신은 생성기**와 같습니다. 그런 다음 **검볼은 임의의 출력**, 즉 얻은 이미지입니다. 어떤 색상의 검볼을 얻을 수 있는지를 통해서, GAN이 무엇에 대해 훈련되었는지 알 수 있게 해줍니다. 하지만, 출력물의 정확한 색상을 제어할 수는 없습니다.
<img src="./images/C1W4.01.png" width="700">
<br>

반면에 **조건적 생성**을 사용하면 특정 클래스에서 예제를 요청하고 요청한 것을 얻을 수 있습니다. **자판기** 같아요. 검볼 기계와 유사하게 동전을 입력하지만, **원하는 항목의 코드와 함께 동전을 입력**합니다. 예를 들어, 레드 소다의 경우 A2입니다. 그러나 여전히 소다병의 특정 특징을 제어할 수 없다는 점에 유의하십시오. 유통 기한이 가장 늦은 것, 가장 덜 손상된 병, 가장 많이 채워진 병을 구할 수 없고 그냥 무작위로 빨간 소다를 얻습니다. 그러나 그것은 파란색 캔디바가 아니라 빨간색 소다입니다. 여기서 **동전과 코드는 GAN에 대한 입력**이고, **자판기는 생성기**이고, **소다는 생성된 출력**입니다. 조건적 GAN을 사용하면 지정한 클래스에서 임의의 예를 얻을 수 있습니다. 그 클래스는 여기 이 A2 소다입니다.

### Conditional vs. Unconditional Generation
이제 조건적 생성과 무조건 생성이 어떻게 관련되어 있는지 알 수 있습니다. 그것들을 조금 비교해봅시다. 조건적 생성을 사용하면 결정한 클래스에서 생성된 예제를 얻을 수 있고, 무조건 생성을 사용하면 임의의 클래스에서 예제를 얻을 수 있습니다. 결과적으로 조건적 생성을 위해서는 실제로 레이블이 지정된 데이터 세트로 GAN을 훈련해야 하며, 이러한 레이블은 원하는 다른 클래스들에 주어져야하며, 무조건 생성에는 레이블이 필요하지 않습니다. 이전 몇 주 동안 이 과정에서 레이블이 필요하지 않고 진짜 예제 더미만 있으면 된다는 것을 보았습니다. 다음 강의에서 이 조건적 생성에 대한 모델을 수정하는 방법을 볼 수 있습니다.
<img src="./images/C1W4.03.png" width="700">
<br>

#### 요약
이 비디오에서 얻어야 할 것은 **조건적 생성에는 원하는 클래스에서 예제를 생성하는 방법을 배우기 위해 훈련을 위해 레이블이 지정된 데이터 세트가 필요하다는 것**입니다. 자판기에서 레드 소다를 선택하는 것과 같이 GAN을 훈련하고 원하는 클래스에서 예제를 생성하기 위해 데이터 세트의 레이블이 생성기와 판별기에 어떻게 공급되는지 보여드리겠습니다.

* 조건적 생성은 라벨이 있는 훈련자료를 요구한다.
* 예제들은 선택된 클래스에 대하여 생성될 수 있다. 

### 조건적 생성 : 입력
조건적 생성을 사용하면 원하는 클래스에서 예제를 생성할 수 있습니다. 그렇게 하려면 레이블된 데이터 세트가 있어야 하고, 훈련 중에 해당 클래스 정보를 생성자와 판별자 모두에 전달해야 합니다. 여기에서는 **생성자에게 생성할 클래스를 알려주는 방법**과 **분류할 이미지와 함께 클래스 정보를 판별자에 통과시키는 방법**을 배웁니다.

* 생성자에게 어떤 종류의 예를 생성하라고 어떻게 말할 수 있는가? 
* 판별자의 입력 표현

#### 생성자 입력
무조건 생성에서 이미 보았듯이 생성기는 임의의 예를 생성하기 위해 노이즈 벡터가 필요합니다. 조건적 생성의 경우 **생성된 예제가 어느 클래스에서 나와야 하는지를 생성기에 알려주는 벡터**도 필요합니다. 일반적으로 이것은 **원-핫(one-hot) 벡터**로, 원하는 클래스에 해당하는 한 위치를 제외한 모든 위치에 0이 있음을 의미합니다. 노이즈 벡터도 임의 값이지만 0 또는 1일 필요는 없습니다. 예를 들어, 원-핫 클래스 벡터의 이러한 모든 다른 값이 서로 다른 견종이라고 가정해 보겠습니다. 여기에서 이 두 번째 요소는 허스키에 해당한다고 합시다. 여기에 1의 위치를 지정한다는 것은 생성기가 허스키를 생성하기를 원한다는 것을 의미합니다. 다른 곳에 1이 있고 허스키 셀에 0이 있는 경우 다른 클래스를 생성하기를 원할 것입니다. 예를 들어 골든 리트리버. 여기서 **노이즈 벡터는 이전과 유사하게 생성에 임의성을 추가하는 벡터**입니다. 다양한 예를 생성할 수 있습니다. 그러나 지금은 특정 클래스에 조건이 지정되고, 두 번째 클래스에 의해 제한되는 특정 클래스 내의 다양한 집합입니다. 이제 생성할 클래스를 제어할 수 있는 one-hot 벡터를 가지게 되었습니다.
<img src="./images/C1W4.04.png" width="700">
<br>

조건적 GAN에서 생성기에 대한 입력은 실제로 **노이즈와 원-핫 클래스 정보가 연결된 벡터**입니다. 따라서 하나의 거대한 벡터입니다. 예를 들어, 여기에 허스키 품종을 나타내는 클래스 벡터가 있는 경우 생성기는 하나의 노이즈 벡터에서 허스키 개를 생성합니다. 
<img src="./images/C1W4.05.png" width="700">
<br>

그러나 **노이즈 벡터를 변경하면 다른 허스키 개를 생성**해야 합니다. 클래스 정보는 동일하게 유지됩니다. 그래서 둘 다 허스키이고, 둘 다 같은 클래스입니다. 그러나 생성기가 실제로 허스키를 생성하는지 어떻게 확인합니까? 이것은 마술처럼 보입니다. 무작위 개를 생성하지 않는 이유는 무엇입니까? 그것은 실제로 판별자에게도 이 클래스 정보가 주어질 것이기 때문입니다. 생성기는 판별자를 속일 필요가 있습니다. 그래서 당신은 이것이 어떻게 일어나는지 보게 될 것입니다.
<img src="./images/C1W4.06.png" width="700">
<br>

유사한 방식으로 판별자는 예제를 사용하지만 이제 예제가 해당 특정 클래스의 진짜 표현인지 가짜 표현인지를 결정하기 위해 입력으로 클래스 정보와 쌍을 이룹니다. 그리고 이것이 여기서 핵심입니다. 예를 들어 이 클래스는 골든 리트리버이고 이미지는 비글견입니다. 따라서 골든 리트리버를 입력으로 주지 않습니다. 이 판별자는 실제로 이것이 가짜라고 말합니다. 비록 그것이 비글의 매우 진짜처럼 보이는 이미지임에도 불구하고, 진짜처럼 보이는 골든 리트리버가 아니기 때문입니다. 이것은 단지 5%정도 진짜라고 할겁니다.
<img src="./images/C1W4.07.png" width="700">
<br>

결과적으로 판별자가 예제를 진짜로 예측하려면, 훈련 데이터 세트에서 해당 클래스의 예제와 같아야 합니다. 훈련 데이터 세트는 옆에 있는 이러한 실제 레이블을 가져오기 때문에 여기에서 핵심입니다. 생성자는 골든 리트리버를 속이기 위해 현실적인 골든 리트리버를 생성해야 합니다. 골든 리트리버 클래스의 이전 예제를 따라서, 판별자는 예제가 골든 리트리버처럼 보이는 경우에만 이미지가 더 진짜임을 예측하는 방법을 학습합니다. 다시 말하지만, 진짜 이미지에 올바른 클래스 레이블이 있고, 이는 생성된 이미지가 올바른 클래스를 사용하도록 권장하기 때문입니다. 꽤 멋지죠?

<img src="./images/C1W4.08.png" width="700">
<br>

조금 더 깊이 파고들면 판별자에 대한 입력은 이미지입니다. 해당 클래스 정보를 추가하는 방법은 무엇입니까? 이미지는 RGB, 빨강, 녹색, 파랑의 세 가지 다른 채널로 입력되거나, 회색조 이미지인 경우 하나의 채널로 입력됩니다. 이것이 바로 이 이미지 컴포넌트입니다. 그런 다음 **원-핫 클래스 정보는 모든 채널이 모두 0 또는 1의 값을 취하는 추가 채널로 제공**될 수도 있습니다. 따라서 이미지와 높이와 너비가 같은 이 모든 흰색 블록은 모두 0의 값을 갖습니다. 반면 이 검은색은 1의 값을 취합니다. 원-핫 벡터와 대조적으로 이들은 일반적으로 각 채널이 해당 클래스가 아닌 모든 위치에서 0으로 가득 찬 훨씬 더 큰 행렬입니다. 이 정보를 다른 형식으로 압축하는 것과 같이 공간을 덜 차지하는 다른 많은 방법이 있습니다. 이를 위해 별도의 신경망 헤드를 생성할 수도 있습니다. 이는 다양한 클래스가 있는 경우에는 신중해야할 것입니다. 그러나 이 접근 방식은 여기에 있는 7개 정도의 클래스에서 확실히 작동합니다. 모델은 해당 정보를 학습합니다.

<img src="./images/C1W4.09.png" width="700">
<br>

#### 요약
조건적 생성을 요약하면 두 모델에 클래스 정보를 전달합니다. **생성기**의 경우 일반적으로 노이즈 벡터와 판별기에 연결된 **원-핫 벡터**이며, **GAN의 원하는 출력이 이미지일 때 판별자**에서는 채널을 나타내는 **원-핫 행렬**입니다. 클래스 **벡터의 크기**와 클래스 정보에 대한 **추가 채널 수**는 제공하는 **클래스 수**와 동일합니다.

* 생성자에게 전달되는 클래스는 one-hot 벡터이다.
* 판별자에게 전달되는 클래스는 one-hot 행렬이다.
* 벡터의 크기와 행렬의 수는 클래스 개수를 나타낸다.

### 제어 가능한 (Controllable) 생성기
GAN에 의해 생성된 출력을 제어하는 또 다른 방법은 훈련이 된 후에 하게 됩니다. 또는 더 일반적으로, **제어 가능한 생성**을 하는 겁니다. 조건적 생성은 훈련 중에 레이블을 활용하지만, 이 비디오는 **모델이 훈련된 후에 출력 예제에서 원하는 특징을 제어**하는 데 중점을 둡니다. 특정 특징을 제어하는 방법과 이전 비디오에서 배운 조건적 생성과 비교하는 방법을 배웁니다.

* 제어 가능한 생성기란?
* 조건적 생성자와의 비교 방법?
<br>

제어 가능한 생성을 사용하면 출력 예제에서 원하는 일부 특징을 제어할 수 있습니다. 예를 들어, 얼굴 생성을 수행하는 GAN을 가지고서, 이미지에서 사람의 나이를 제어하거나, 선글라스를 착용하거나, 사진에서 바라보는 방향을 제어하거나, 인식되는 성별을 제어할 수 있습니다.

<br>

모델을 훈련시킨 후 생성기에 공급되는 입력 노이즈 벡터 Z를 실제로 조정하여 이를 수행할 수 있습니다. 예를 들어 여기 입력 노이즈 벡터 Z를 사용하면 빨간 머리를 한 여성의 사진을 얻을 수 있습니다.<img src="./images/C1W4.11.png" width="700">

<br>

그러나 먼저 제어 가능한(controllable) 생성을 더 잘 이해하기 위해 조건적(conditional) 생성과 빠르게 비교하겠습니다. 나는 여기에서 이 용어를 사용할 것입니다. 왜냐하면 그것이 연구자들이 일반적으로 논문에서 사용할 때 의미하는 것이기 때문입니다. 비록 명확하게 구분되지는 않았지만 말입니다. 때때로 제어 가능한 생성은 여전히 어떤 식으로든 GAN을 제어하고 있기 때문에 조건적 생성을 확실히 포함할 수 있습니다. 제어 가능한 생성을 사용하면 녹색 머리와 안경을 쓴, 나이 들어 보이는 사람들의 얼굴과 같이 원하는 특징을 가진 예를 얻을 수 있습니다. 조건적 생성을 사용하면 인간이나 새와 같이 원하는 클래스에서 예제를 얻을 수 있습니다. 물론 그럴 수도 있겠지만, 선글라스를 쓰고 있는 사람을 원할 수 도 있습니다. 여기까지는 약간 비슷합니다. 
그러나 **제어 가능한 생성은 일반적으로 원하는 특징의 양의 많고 적음을 제어하려는 것을 의미**합니다. 일반적으로 연령과 같은 **보다 연속적인 특징**입니다. 
반면에 **조건적 생성을 사용하면 매우 다른 유형의 사물에 대해 원하는 클래스를 지정**할 수 있습니다. 이를 위해서는 일반적으로 훈련 중에 레이블 데이터 세트와 함께하는 구현이 필요합니다. 
모든 머리카락 길이 값에 레이블을 지정하고 싶지 않을 수 있으므로, 제어 가능한 생성이 이를 수행하고, 여기에서는 원하는 특징의 방향을 찾는 것이 더 중요하며, 이는 훈련 후에 일어납니다. 
물론 제어 가능한 생성이 훈련 중에도 발생하는 것을 볼 수 있습니다. 제어하기 쉬운 방향으로 모델을 이동하는 데 도움이 됩니다. 마지막으로, 방금 배운 것처럼 제어 가능한 생성은 생성기에 공급되는 입력 노이즈 벡터 Z를 조정하여 작동하지만, 조건적 생성에서는 해당 노이즈 벡터에 추가하려는 클래스를 나타내는 추가 정보를 전달해야 합니다.

<img src="./images/C1W4.13.png" width="700">
<br>


#### 요약
요약하자면. 제어 가능한 생성을 통해 GAN 출력의 특징을 제어할 수 있습니다. 대조적으로, 제어 가능한 생성에서는 레이블이 지정된 훈련 데이터 세트가 필요하지 않습니다. 제어 가능한 생성 방식으로 출력을 변경하려면 입력 노이즈 벡터를 다른 방식으로 조정합니다. 다음 동영상에서 정확히 어떻게 작동하는지 자세히 알아보겠습니다.

* 제어 가능한 생성자는 생성된 출력의 특징을 제어할 수 있게 해준다
* 이는 라벨된 훈련자료를 요구하지 않는다.
* 다른 특징의 출력을 얻기위해서 입력 벡터를 조정한다. 

### Z-공간에서의 벡터 대수
이전 비디오에서 보았듯이 제어 가능한 생성은 생성기에 공급되는 노이즈 벡터 z를 조작하여 달성됩니다. 이 비디오에서는 그 과정 뒤에 숨겨진 직관을 보여 드리겠습니다. 먼저 두 GAN 출력 사이를 보간하는 방법을 검토하고, 출력에서 원하는 특징을 제어하기 위해 노이즈 벡터를 조작하는 방법을 배우게 됩니다.

* Z-공간에서의 내삽(interpolation)
* 원하는 특징의 조정을 위한 잡음벡터 z의 수정 
<br>

제어 가능한 생성과 보간은 다소 비슷합니다. 보간법을 사용하면 생성된 두 관측값 사이에 이러한 중간 예를 얻을 수 있습니다. 실제로 보간을 사용하면 이 GIF에서와 같이 이미지가 다른 이미지로 어떻게 변형되는지 확인할 수 있습니다. 여기서 0에서 9까지의 각 숫자는 다음 숫자로 변형됩니다. 정말 멋지네요.
<img src="./images/C1W4.14.png" width="700">
<br>

노이즈 벡터의 벡터 공간에 대한 이름인 $Z$-공간의 입력을 조작하여, 대상 간에 중간 예를 얻을 수 있습니다. 이것이 제어 가능한 생성 이면의 동일한 아이디어라는 것을 나중에 알게 될 것입니다. 여기에서 명확히 하자면 $z_1$과 $z_2$는 지금 보고 있는 이 $Z$-공간의 두 차원입니다. 예를 들어, 노이즈 벡터 $v_1$과 노이즈 벡터 $v_2$가 있습니다. 여기서 $v_1$은 $z_1$ 값이 5이고 $z_2$ 값이 10이라고 가정해 보겠습니다. 그러면 이것은 벡터 $[5, 10]$이고 $v_2$는 더 작은 값을 갖습니다. 그래서 값이 4와 2. 그래서 그것은 벡터 $[4, 2]$입니다. 
따라서 $z_1$과 $z_2$는 $Z$-공간의 각 차원이며, 실제 벡터 $v_1$과 $v_2$가 이 $Z$-공간의 구체적인 벡터 값을 나타낼 것입니다. $v_1$은 생성기에 공급하면 여기에서 이 이미지를 생성하고 $v_2$는 생성기에 공급할 때 이 이미지를 생성합니다.
<img src="./images/C1W4.15.png" width="700">
<br>

이 두 이미지 사이의 중간 값을 얻으려면, 실제로 $Z$-공간에서 두 입력 벡터 $v_1$ 및 $v_2$ 사이를 보간할 수 있습니다. 이 보간은 종종 선형 보간입니다. 물론, 이 두 벡터 사이를 보간하는 다른 방법이 있습니다. 그런 다음 이 모든 중간 벡터를 가져와서 생성기에서 생성하는 것을 볼 수 있습니다. 생성기는 이 벡터를 가져와서 해당 이미지, 이 벡터는 이 이미지, 이 벡터는 이 이미지를 생성하여 이 두 이미지 사이의 점진적 변화를 얻습니다.
<img src="./images/C1W4.16.png" width="700">
<br>

제어 가능한 생성은 $Z$-공간의 변경 사항을 사용하며, 잡음 벡터에 대한 변화가 생성기의 출력에 반영되는 방식을 활용합니다. 예를 들어 한 노이즈 벡터를 사용하면 빨간 머리를 한 여성의 사진을 얻을 수 있고, 다른 노이즈 벡터를 사용하면 같은 여성이지만 파란 머리를 가진 사진을 얻을 수 있습니다. 이 두 노이즈 벡터의 차이점은 생성된 이미지의 머리색을 수정하기 위해 $Z$-공간에서 이동해야 하는 방향입니다. 제어 가능한 생성에서 목표는 관심 있는 다양한 특징에 대한 이러한 방향을 찾는 것입니다. 예를 들어, 머리 색깔 수정. 하지만 아직 정확한 방향을 찾는 것에 대해 걱정하지 마십시오. 다음 강의에서 보여드리겠습니다. 이 알려진 방향 $d$를 사용하여(이 방향을 $d$라고 합시다), $Z$-공간에서 이제 GAN 출력의 기능을 제어할 수 있습니다. 이는 정말 흥미로운 일입니다.
<img src="./images/C1W4.17.png" width="700">
<br>

즉, 동일한 생성기 $G$에서 생성된 빨간 머리를 가진 남자의 이미지를 생성하고, 여기에 이 입력 노이즈 벡터 $v_1$을 사용하면 해당 방향 벡터 $d$를 추가하여 이미지에서 이 남자의 머리 색상을 수정할 수 있습니다. 노이즈 벡터 이전에 발견된 이 새로운 노이즈 벡터 $v_1 + d$를 생성하여 생성기로 전달하고 머리카락이 이제 파란색인 결과 이미지를 얻습니다. 
<img src="./images/C1W4.18.png" width="700">
<br>


#### 요약
요약하자면, 제어 가능한 생성에서는 GAN의 출력에서 원하는 특징의 변경과 관련된 Z 공간의 방향을 찾아야 합니다. 알려진 방향으로 제어 가능한 생성은 해당 $Z$ 공간에서 노이즈 벡터를 다른 방향으로 이동하여 작동합니다. 다음으로, 제어 가능한 생성과 관련된 몇 가지 문제와 생성된 출력에 알려진 효과가 있는 $Z$-공간의 방향을 찾는 방법을 배우게 됩니다.

* 출력 특징을 조정하기 위해서는 Z-공간에서의 방향을 찾을 필요가 있다.
* 출력을 변경하기 위해서는 Z-공간에서 이동해야 합니다.

### 제어가능한 생성자의 과제
제어 가능한 생성은 사람의 사진을 생성하는 GAN에서 머리 색깔이나 머리 길이와 같은 GAN의 특징과 출력을 결정할 수 있도록 합니다. 그러나 제어 가능한 생성에는 이 비디오에서 볼 수 있는 몇 가지 문제가 있습니다. 특히 **특징 상관** 및 **출력 공간과 Z-공간의 얽힘**에 대해 학습합니다.

* 출력 특징의 상관관계
* Z-공간의 얽힘(entanglement)

#### 특징 상관(Feature Correlation)
**서로 다른 특징이 데이터 세트에서 높은 상관 관계를 갖는 경우**, GAN을 훈련하는 데 사용하면 상관 관계가 있는 다른 특징을 수정하지 않고 특정 특징을 제어하기가 어려워집니다. 예를 들어, 이상적으로는 GAN으로 생성된 사람의 얼굴 수염 양과 같은 단일 특성을 제어할 수 있고, 데이터 세트의 특성이 높은 상관 관계를 갖지 않는 경우 다음을 수행할 수 있습니다. Z-공간에서 어떤 방향으로 이동하여 여성의 이 사진에 수염을 추가합니다. 그러나 데이터 세트에서 수염의 존재와 얼굴이 얼마나 남성적으로 보이는 지와 같은 특성을 훈련하는 데 사용하는 것은 강한 상관관계가 있을 가능성이 매우 높습니다. 여성 사진에 수염을 추가하려면 출력에서 더 많은 특성들이 수정되게 됩니다. 그러나 누군가에 대한 한 가지 특성만 변경할 수 있는 이러한 지침을 찾을 수 있기를 원하기 때문에 바람직하지 않을 수 있습니다. 그렇게 하면 이미지를 안정적으로 편집할 수 있습니다.
<img src="./images/C1W4.19.png" width="700">
<br>

#### Z-공간 얽힘
제어 가능한 생성이 직면한 또 다른 문제는 Z-공간의 얽힘으로 알려져 있습니다. Z-공간이 얽힌 경우, **특정 방향으로의 잡음 벡터 이동은 출력에서 동시에 여러 특징에 영향**을 줍니다. 이러한 특징이 훈련 데이터 세트에서 반드시 상관관계가 있는 것은 아닐지라도. 이것이 바로 **잡음 공간이 학습된 방법**이며, 이는 매우 얽혀 있을 수 있습니다.
<img src="./images/C1W4.20.png" width="700">
<br>

이 얽힌 Z-공간에서 출력에 있는 사람이 안경을 쓰고 있는지 여부를 제어할 때, 예를 들어 머리카락에 수염이 있는지 여부도 수정하게 되거나, 
<img src="./images/C1W4.21.png" width="700">
<br>

겉보기 나이를 수정하려고 할 때도 그녀의 명백한 눈과 머리 색깔 결과도 수정됩니다. 그다지 바람직하지 않습니다. 다른 상관되지 않은 특징에서도 마찬가지입니다. 
<img src="./images/C1W4.22.png" width="700">
<br>

이는 노이즈 벡터의 일부 구성 요소가 변경되면 출력의 여러 특징이 동시에 변경되고 이로 인해 출력을 제어하는 것이 불가능하지는 않더라도 어렵습니다. 이것은 **Z 공간이 출력에서 제어하려는 특징의 수에 비해 충분한 차원을 갖지 않을 때** 매우 일반적인 문제입니다. 그러면 실제로 일대일로 매핑할 수 없기 때문입니다. 이것은 생성 모델을 훈련할 때 일반적으로 발생하는 문제이기도 합니다.
<img src="./images/C1W4.23.png" width="700">
<br>


#### 요약
요약하자면, 제어 가능한 생성은 몇 가지 문제에 직면해 있습니다. 데이터 세트의 특징이 서로 높은 상관 관계를 갖고 있고, 이를 어떤 식으로든 설명하지 않으면 GAN의 출력을 제어하려고 할 때 결국 여러 특징을 변경하게 됩니다. 제어하고자 하는 특징이 훈련 데이터 세트에서 서로 높은 상관관계가 없더라도, Z-공간이 얽혀 있으면 생성 제어도 어렵습니다. Z-공간의 차원 수가 충분히 크지 않은 경우 일반적으로 발생하지만 그러한 일이 발생하는 다른 이유도 있습니다.

* 하나의 특징을 조정하고자 한다면, 상관된 다른 것들도 변화한다.
* Z-공간 얽힘이 제어를, 불가능하지 않다면, 어렵게 한다.
* 얽힘은 Z-공간의 차원이 충분하지 않을 때 발생한다.

### 분류기 그래디언트(Classifier Gradients)
제어 가능한 생성은 머리카락을 늘리거나 줄이는 것과 같이 원하는 특징에 해당하는 방향에 따라 z 공간에서 이동하여 작동합니다. 이 비디오에서는 훈련된 분류기의 기울기를 사용하여 그 방향을 찾는 데 사용되는 매우 간단한 방법을 배우고, 이 방법에 대한 요구 사항을 확인할 수 있습니다.

* Z-공간에서 방향을 찾기 위해 분류기를 어떻게 사용하는가?
* 이 방법을 사용하는데 있어서 요구조건
<br>

출력의 특정 특징을, 즉 선글라스의 존재, 수정하는 z-공간의 방향을 찾으려 합니다. 이를 위해서 사진 속 인물이 해당 특징을 가지고 있는지 식별하는 훈련된 분류기를 사용할 수 있습니다. 이렇게 하려면 생성기를 통과하여 일부 이미지를 얻는 일련의 노이즈 벡터 Z를 사용합니다. 그런 다음 이 이미지를 선글라스 분류기를 통해 전달하면 선글라스가 있거나 없는 사람에 해당하는 출력을 알려줍니다. 그런 다음 그 정보를 사용하여 Z 벡터를 수정합니다. 이것은 생성기의 가중치는 전혀 수정하지 않는 것입니다. 따라서 생성기의 가중치는 고정됩니다. 훈련은 이미 마쳤습니다. 
대신에 해당 비용의 그래디언트 방향으로 이동하도록 Z 벡터를 수정합니다. 이는 선글라스를 착용하지 않은 것으로 분류된 모든 이미지에 대해 모델에게 불이익을 줍니다.

따라서 이 방법은 매우 간단하며, 효율적입니다. 일부에서는 이미 있는 분류기를 사용하고 있기 때문에 게으르다고 주장할 수도 있습니다. 그러나 사전 훈련 분류기를 활용하고 있다는 점은 훌륭하다고 생각합니다. 그러나, 물론 그것의 단점도 있습니다. 제어하려는 특징을 정확하게 감지하는 사전 훈련된 분류기가 필요합니다. 물론 직접 훈련할 수도 있으므로, 예를 들어 수염을 감지한 분류기를 사용하려는 경우, 사용할 수 있는 분류기가 없는 경우, 스스로 훈련해야 할 수도 있습니다. 외부에 무엇이 있는지 항상 확인해야 합니다. 이것은 제어를 시작할 수 있는 정말 간단하고 멋진 방법일 수 있기 때문입니다.

<br>

#### 요약
따라서 이 비디오에서 사전-훈련된 분류기를 사용하여 출력의 특징과 관련된 z-공간의 방향을 찾을 수 있다는 점을 알아가야 합니다. 그리고 분류기의 그라디언트를 사용하여 이러한 방향을 찾으려면 생성기를 변경하지 않고 노이즈 벡터를 수정해야 합니다. 그러므로 이 모든 것은 훈련 후에 일어난다는 것을 기억하십시오.

* 분류기가 Z-공간에서의 방향을 찾는데 사용될 수 있다.
* 방향을 찾기위해서, 단지 잡음 벡터에만 개선하면 된다.

### 얽힘 풀림(disentanglement)
이전에 $z$-공간 얽힘이 의미하는 것과 이것이 제어 가능한 생성에서 문제가 되는 이유를 보았습니다. 이 비디오에서는 얽힘 풀림과 모델이 이를 달성하도록 권장하는 몇 가지 방법에 대해 이야기하겠습니다. 먼저 얽힘 대 얽힘-풀린 z 공간이 의미하는 바를 다시 살펴본 다음, 모델이 얽힘-풀린 z 공간를 갖도록 권장하는 가장 인기 있는 몇 가지 방법을 언급하겠습니다.

* 얽힘-풀린 Z-공간은 무엇을 의미하는가?
* 얽힘-풀린 Z-공간을 장려하는 방법들
<br>

다음 $z$-노이즈 벡터 $v_1$ 및 $v_2$를 사용하고, 이들이 얽힘 풀린 $z$-공간에서 나온다고 상상해 보십시오. 이제 여기 오른쪽에서 $z$-공간의 2차원을 시각화하고 있습니다. 하지만 물론 이러한 노이즈 벡터에서 알 수 있듯이 2차원보다 더 큰 차원의 것들 입니다. 이제 이것이 얽힘 풀린 $z$-공간이면 이러한 각 위치는 출력의 단일 특징에 해당합니다. 예를 들어, 이 첫 번째 요소, 노이즈 벡터의 첫 번째 차원은 머리카락 색에 해당하고, 두 번째 차원은 머리카락 길이에 해당하는 식입니다. 그림에서 머리카락 색을 변경하려면 말 그대로 노이즈 벡터의 첫 번째 요소를 변경하거나, 해당 $z$-공간에서 $z_1$ 방향으로 이동하면 됩니다. 머리 색깔을 바꾸기만 하면 됩니다. 그런 다음 머리카락 길이의 경우 $z_2$ 차원으로 이동하면 됩니다. 그것은 바로 여기 두 번째 위치에서 이 값을 변경하는 것입니다. 


그리고, 이러한 노이즈 벡터가 이 두 가지 다른 특징과 관련하여 얽힘이 풀려있고 나머지 값은 아무 의미가 없을 수도 있습니다. 이는 일반적으로 노이즈 벡터의 크기가 제어하려는 특징의 수보다 크게 되기를 원하기 때문입니다. 이렇게 하면 훈련 중 시간이 지남에 따라 이러한 값이 다른 것을 취할 수 있기 때문에 모델이 더 쉽게 학습할 수 있습니다. 예를 들어 머리카락 색과 머리카락 길이와 같은 두 가지 특징만 제어하려는 경우 노이즈 벡터를 2차원보다 크게 만드는 것이 현명할 수 있으며, 여기서는 3개 이상이 표시됩니다. 이러한 다른 값은 특정 특징을 제어하지 않으며 모델이 훈련에 적응하는 데 도움이 될 뿐입니다. 


얽힘 풀린 z 공간에 있는 노이즈 벡터의 구성 요소를 사용하면 출력에서 원하는 특징을 변경할 수 있기 때문에 이를 종종 변동의 잠재 요인(latent factors of variation)이라고 하며, 여기서 잠재(latent)라는 단어는 노이즈의 정보가 벡터는 출력에서 직접 볼 수는 없지만 출력이 어떻게 보이는지 결정한다는 뜻입니다. 때때로 노이즈 벡터라고 부르기 보다는 일반적으로 잠재성(latents)이라고 하는 것을 들을 수 있습니다. 그런 다음 변형 요인(factors of variation)이란 머리 색깔과 머리 길이와 같은 다양하게 만들고자 하는 서로 다른 요소들을 말하며, 변화시킬 때 달라지는 것은 한 가지 요소, 즉, 단 하나의 특징라는 것을 의미합니다.
<img src="./images/C1W4.27.png" width="700">
<br>

본질적으로 얽힘-풀린 z 공간은 노이즈 벡터에 특정 인덱스가 있음을 의미합니다. 이러한 특정 차원은 GAN 출력의 특정 특징을 변경합니다. 예를 들어, GAN에 의해 생성된 이미지의 사람이 안경을 쓰고 있는지 여부, 수염이 있는지 또는 머리카락의 일부 특징이 있는지 여부. 이러한 각 셀은 변경하려는 항목에 해당하며 안경, 수염 또는 머리카락을 적용하기 위해 해당 치수의 값을 변경할 수 있습니다. 
<img src="./images/C1W4.28.png" width="700">
<br>

얽힘-풀린 z-공간과 얽힌 z-공간의 중요한 차이점은 여기에서 얽힘-풀린 z-공간을 사용하여 출력의 특징(예: 안경) 중 하나를 제어할 때 다른 특징이 동일하게 유지된다는 것입니다. 수염과 머리카락은 그대로 유지됩니다. 또는 누군가 수염이 있는지 여부를 변경하면 안경과 머리카락은 그대로 유지됩니다. 
이것이 의미하는 바는 얽힘-풀린 z-공간을 사용하면 머리카락이나 얼굴 특징을 바꾸지 않고도 매우 여성스러워 보이는 사람에게 수염을 추가할 가능성이 훨씬 더 높다는 것입니다.
<img src="./images/C1W4.29.png" width="700">
<br>

모델이 얽힘-풀린 z-공간을 사용하는 하도록 독려하는 방법은 데이터에 레이블을 지정하고 조건적 생성에 사용되는 유사한 프로세스를 따르는 것입니다. 
<img src="./images/C1W4.30.png" width="700">
<br>

그러나 이 경우 클래스의 정보가 노이즈 벡터에 포함됩니다. 따라서 이 추가 원핫 클래스 정보나 클래스 벡터가 필요하지 않습니다. 그러나 이 방법을 사용하면 연속 클래스에 문제가 될 수 있습니다. 수천 명의 사람의 얼굴에 머리카락 길이를 표시해야 한다고 상상해 보십시오. 물론 이렇게 하면 몇 가지 다른 클래스, 몇 가지 다른 버킷이 있어도 실제로 생성기를 올바른 방향으로 움직일 수 있습니다.<img src="./images/C1W4.31.png" width="700">
<br>

모델에 장려되는 또 다른 방법은 예제에 레이블을 지정하지 않고 얽힘-풀린 z 공간을 사용하는 것입니다. 대신 BCE 또는 W-loss와 같이 선택한 손실 함수에 정규화 항을 추가하여 모델이 출력의 두 가지 다른 특징인 노이즈 벡터의 각 인덱스를 연결하도록 권장합니다. 이 정규화는 분류기 그라디언트(classifier gradient)에서 올 수 있으며, 레이블이 없는 비지도(unsupervised) 방식으로 이를 수행하는 훨씬 더 고급 기술도 있습니다.

<br>

#### 요약
요약하면, 얽힘 풀린 z-공간을 사용하면 제어하려는 원하는 특징에 특정 z-값을 직접 대응시켜 개별 출력 특징을 제어할 수 있습니다. 모델이 얽힘 풀린 노이즈 벡터를 사용하도록 장려하려면 지도 학습 방법과 비지도 학습 방법을 모두 사용할 수 있습니다.

* 얽힘 풀린 Z-공간은 각각의 특징들에 해당하는 z-값을 직접 조정할 수 있게 해준다.
* 얽힘 풀림를 이루기 위해서는 지도 및 비지도 방법이 있다.
